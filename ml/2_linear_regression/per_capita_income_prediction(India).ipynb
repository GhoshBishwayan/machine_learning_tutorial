```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('india_income.csv')
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita inocme')
plt.scatter(df.year,df.per_capita_income,color='green')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.per_capita_income)
```

```python
reg.predict(pd.DataFrame({'year':[2026]}))
```

```python
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
y = m*2026+b
y
```

```python
%matplotlib inline
plt.scatter(df.year,df.per_capita_income,color="green")
# plt.plot(df[['year']],df.per_capita_income,color='blue')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
