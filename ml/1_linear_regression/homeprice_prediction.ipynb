```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('homeprices.csv')
```

```python
%matplotlib inline
plt.xlabel('area')
plt.xlabel('price')
plt.scatter(df.area,df.price,color='green')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price) #fit means training the data
#Here it's based on area predict price
```

```python
x = 3100
reg.predict(pd.DataFrame({'area':[x]}))
```

```python
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
y = m*x+b
y
```

```python
%matplotlib inline
plt.xlabel('area')
plt.xlabel('price')
plt.scatter(df.area,df.price,color='green')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
